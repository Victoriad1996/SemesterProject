```python
import matplotlib.pyplot as plt
from brian2 import*
```

```python
def visualise_connectivity(S, mylegend):
    Ns = len(S.source)
    Nt = len(S.target)
    figure(figsize=(10, 4))
    subplot(121)
    plot(zeros(Ns), arange(Ns), 'ok', ms=10)
    plot(ones(Nt), arange(Nt), 'ok', ms=10)
    for i, j in zip(S.i, S.j):
        plot([0, 1], [i, j], '-k', linewidth=S.w[i,j][0])
    xticks([0, 1], ['Source', 'Target'])
    ylabel('Neuron index')
    xlim(-0.1, 1.1)
    ylim(-1, max(Ns, Nt))
    subplot(122)
    plot(S.i, S.j, 'ok')
    xlim(-1, Ns)
    ylim(-1, Nt)
    xlabel('Source neuron index')
    ylabel('Target neuron index')
    suptitle(mylegend)
```

```python
start_scope()

#Recurrent weight from PC to PC
w_PC = 0.5
#Weight PC to INH
w_PCINH = 0.7
#Weight INH to PC
w_INHPC = 0.5
#Weight G to PC
w_GPC = 0.9
#Weigth S to PC
w_SPC = 0.8

eqs = '''
dv/dt = (I-v)/tau : 1
I : 1
tau : second
x : metre
w : 1
g : 1
'''

N = 30
neuron_spacing = 50*umetre
width = N/4.0*neuron_spacing

# Pyramidal cells
PC = NeuronGroup(N, eqs,  threshold='v>0.8', reset='v = 0.1', method='exact')
PC.I = '0'
PC.tau = '10*ms'
PC.g = 1


# All synapses are connected (excluding self-connections)
SPC = Synapses(PC, PC, on_pre='v_post += 0.01')
SPC.connect(condition='i!=j')
# Weight varies with distance
SPC.w = 'w_PC*exp(-(x_pre-x_post)**2/(2*width**2))'

#Inhibitory cells
ninh = 3
p = 0.5
IN = NeuronGroup(ninh, eqs,  threshold='v>0.8', reset='v = 0.', method='exact')
IN.I = '0.9'
IN.tau = '10*ms'

#Connect INH to PC with probability 1/2
SIPC = Synapses(IN, PC, on_pre= 'v_post -= 0.1 * v_pre')
SIPC.connect(p = 0.5)
SIPC.w = 'w_INHPC'

#Connect PC to Inh 
SPCI = Synapses(PC, IN, on_pre= 'v_post += 0.002')
SPCI.connect()
SIPC.w = 'w_INHPC'

#G inputs
nG = N
p = 0.6
G =  NeuronGroup(nG,  eqs,  threshold='v>0.8', reset='v = 0', method='exact')
G.I = 1
G.tau = '10*ms'

#Connect G inputs to PC
SGPC = Synapses(G, PC, on_pre='v_post += 0.01')
SGPC.connect(p = p)
SGPC.w = w_GPC


#S inputs
nS = 30
S =  NeuronGroup(nS,  eqs,  threshold='v>0.8', reset='v = 0', method='exact')
S.I = 1.5
S.tau = '10*ms'

#Connect G inputs to PC
SSPC = Synapses(S, PC, on_pre='v_post += 0.5')
SSPC.connect(condition = 'i==j')
SSPC.w = w_GPC
```

```python

MPC = StateMonitor(PC, 'v', record=0)
MG = StateMonitor(G, 'v', record=0)
MI = StateMonitor(IN, 'v', record=0)
MS = StateMonitor(S, 'v', record=0)

run(50*ms)

printPC = True 

if printPC:
    linePC, = plot(MPC.t/ms, MPC.v[0], label='PC')
    lineI, = plot(MI.t/ms, MI.v[0], label='I')
    lineG, = plot(MG.t/ms, MG.v[0], label='G')
    lineS, = plot(MS.t/ms, MS.v[0], label='S')
    legend(handles = [ linePC,lineI, lineG, lineS])
else:
    linePC, = plot(MPC.t/ms, MPC.v[0], label='PC')
    lineS, = plot(MS.t/ms, MS.v[0], label='S')
    legend(handles = [linePC, lineS])

xlabel('Time (ms)')
ylabel('v')
```

```python
#Visualise


#Visualise recurrent connectivity
visualise_connectivity(SPC, "Excitatory")

#Visualise inhibitory connectivity
visualise_connectivity(SIPC, "Inhibitory connected to PC")

#Visualise G connectivity
visualise_connectivity(SGPC, "G connected to PC")


#Visualise S connectivity
visualise_connectivity(SSPC, "S connected to PC")
```
